```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
```

```python
# Read the dataset
TitanicPassengers = pd.read_csv("https://raw.githubusercontent.com/SandrineLpx/510B_class_exercices/refs/heads/main/TitanicPassengers.csv")
print(TitanicPassengers.describe())
```

```python
df_train, df_test = train_test_split(TitanicPassengers, test_size=0.4, random_state=42)
```

```python
logit_regression = smf.logit("Survived ~ Class + Age + Sex", data=df_train)
logit_results    = logit_regression.fit()
print(logit_results.summary())
```

```python
df_test['PredictedSurvived'] = logit_results.predict(df_test) > 0.5
```

```python
pd.crosstab(df_test['Survived'], df_test['PredictedSurvived'])
```
